```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,7,4]
plt.plot(x, y, label = 'First line', marker='o', markerfacecolor='r')
y2 = [10,11,5]
plt.plot(x, y2, label = 'Second line', marker='o', markerfacecolor='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('My Graph')
plt.legend()
plt.show()
```

```python
import numpy as np
xvals = np.arange(-2, 10, 0.01)
yvals = np.sin(xvals)
plt.plot(xvals, yvals)
plt.show()
```

```python
def fnplot(list1):
    plt.plot(list1, marker='o', markerfacecolor='r')
    plt.title("Marks Chart")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
list1 = [50,50,50,65,65,75,75,80,80,90,90,90,90]
fnplot(list1)
```

```python
import numpy as np
x = np.arange(12, 21)
y = 10*x + 14
plt.title("Algebraic Expr")
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.plot(x, y, color='r', linestyle='dashed', marker='o', markerfacecolor='navy')
plt.show()
```

```python
import numpy as np
city = ['Delhi', 'Mumbai', 'Chennai', 'Pune']
p = [1500,2000,1200,1800]
plt.bar(city, p, color=['gold', 'orange', 'orangered', 'red'])
plt.xlabel("City")
plt.ylabel("Population in Lacs")
plt.grid(True)
plt.grid(color='k')
plt.show()
```

```python
import numpy as np
obj = ('Python', 'C++', 'Java', 'Perl', 'Scala')
y = np.arange(len(obj))
performance = [10,8,7,4,2]
plt.barh(y, performance, align='center', color=['r','brown','orange','gold','chocolate'])
plt.yticks(y,obj)
plt.xlabel("Usage")
plt.show()
```

```python
import numpy as np
label = ['Anil', 'Vikas', 'Dharma', "Manish"]
per = [94,85,45,50]
index = np.arange(len(label))
plt.bar(index, per)
plt.xlabel('Student name', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.xticks(index, label, fontsize=12, rotation=30)
plt.title("Percentage of marks")
plt.show()
```

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes  =[15,30,45,20]
explode = (0, 0.1, 0, 0)
fig1, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=50)
ax.axis('equal')
plt.show()
```
